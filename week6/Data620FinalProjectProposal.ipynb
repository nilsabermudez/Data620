## Proposal for Final Project

### Question

How has election-related misinformation spread through social media networks during the 2020 and 2024 U.S. elections, and what can sentiment analysis reveal about public engagement and discourse around these claims?

### Motivation

Understanding how misinformation spreads is critical for improving public trust, informing policy decisions, and designing interventions. By analyzing network dynamics and sentiment around election-related claims, we aim to uncover patterns in how misinformation circulates and evolves over time.

### Dataset(s)

- 2020 election claims on Twitter: https://www.kaggle.com/datasets/paultimothymooney/voterfraud2020-election-fraud-claims-on-twitter/data
- 2024 election sentiment on X (Twitter): https://www.kaggle.com/datasets/emirhanai/2024-u-s-election-sentiment-on-x
- Additional reference: https://arxiv.org/html/2407.16051v1#S4

#### Data Preparation & Exploration

- Load, clean, and normalize datas+ets.
- Identify and extract key features: user accounts, retweets, replies, hashtags, and claims.

#### NLP Pipeline

- Extract prominent claims or topics from text using NLP techniques (e.g., topic modeling, keyword extraction).
- Conduct sentiment analysis to assess public emotions and attitudes toward different claims.
- If feasible, cross-reference claims with fact-checking sources to label possible misinformation (acknowledging this as a challenge).

#### Network Analysis & Visualization

- Construct interaction networks (using NetworkX or Plotly) showing connections among accounts, retweets, and shared content.
- Visualize network structures to highlight influential accounts, communities, and pathways of misinformation spread.

### Concerns & Challenges

- Temporal inconsistency (2020 vs. 2024 data) may complicate direct comparison, so we will focus on thematic analysis instead.
- Variations in data labeling, formats, and metadata may require additional preprocessing and harmonization.
- Limited access to reliable fact-checking data could constrain our ability to formally label misinformation.
- Potential class imbalance in misinformation-related labels might affect model evaluation and interpretation.

### Expected Outcomes

- A reproducible workflow (in IPython Notebook or R Markdown) detailing our analysis from raw data to insights.
- Visualizations illustrating network structures and misinformation pathways.
- Comparative sentiment analysis of election-related discourse in 2020 and 2024.
- Insights into key actors and communities involved in misinformation spread.
