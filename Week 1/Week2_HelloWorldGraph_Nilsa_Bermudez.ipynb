```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G = nx.krackhardt_kite_graph()

# Use a layout
pos = nx.circular_layout(G)

# Draw the graph
plt.figure(figsize=(8, 6))
# Mapping from numbers to names
name_mapping = {
    0: "Andre",
    1: "Beverly",
    2: "Carol",
    3: "Diane",
    4: "Ed",
    5: "Fernando",
    6: "Garth",
    7: "Heather",
    8: "Ike",
    9: "Jane"
}

# Step 3: Apply the mapping to relabel the graph
G = nx.relabel_nodes(G, name_mapping)

# Step 4: Draw the graph with a layout
pos = nx.kamada_kawai_layout(G)  

plt.figure(figsize=(8, 6))
nx.draw(G, pos=pos, with_labels=True, node_color='skyblue', node_size=1000, font_weight='bold')
plt.title("Nilsa Bermudez - Krackhardt Kite Graph with Named Nodes")
plt.show()
```
